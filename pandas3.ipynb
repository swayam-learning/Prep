```python
df = pd.DataFrame([
    [(3,4),(-3,4)],
    [(-6,5),(0,0)],
    [(0,0),(2,2)],
    [(-10,1),(10,10)],
    [(4,5),(1,1)]
    
    ],columns=["1st_point","2nd_point"])
```

```python
def euclidean(row):
    pt_A = row["1st_point"]
    pt_B = row["2nd_point"]
    return ((pt_A[0]-pt_B[0])**2+(pt_A[1]-pt_B[1]**2)**0.5)
```

```python
df["distance"] = df.apply(euclidean,axis=1)
df
```
